```python
import numpy as np
import matplotlib.pyplot as plt

def cef(x):            # given CEF
    return (3*x**2 + 2) / (6*x**2 + 3)

def linear_fit(x):     # from part (a)
    return 96/151 - (10/151)*x

x = np.linspace(0, 2, 400)
plt.figure(figsize=(8,5))
plt.plot(x, cef(x), label="CEF: $E[Y|X=x]=\\frac{3x^2+2}{6x^2+3}$", linewidth=2)
plt.plot(x, linear_fit(x), "--", label="Best linear fit $\\ell(x)=\\frac{96}{151}-\\frac{10}{151}x$", linewidth=2)
plt.xlabel("x"); plt.ylabel("y"); plt.title("CEF vs Best Linear Fit on $[0,2]$")
plt.legend(); plt.grid(True); plt.tight_layout()
plt.show()
```
